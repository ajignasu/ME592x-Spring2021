```python
#### Naive learning and LSTM

import os
import torch
import torch.nn as nn
import torch.nn.functional as F

def reset_weights(m):
  '''
    Try resetting model weights to avoid
    weight leakage.
  '''
  for layer in m.children():
    if hasattr(layer, 'reset_parameters'):
      print(f'Reset trainable parameters of layer = {layer}')
      layer.reset_parameters()


class Fully_Connected(nn.Module):
    def __init__(self, in_dims, out_dims):
        super(Fully_Connected, self).__init__()
        
        self.layer_1 = nn.Linear(in_dims, 250) # fully-connected net
        self.layer_2 = nn.Linear(250, 125) # fully-connected net
        self.layer_3 = nn.Linear(125, 50) # fully-connected net
        self.layer_4 = nn.Linear(50, out_dims) 

    def forward(self, x):
        x = F.relu(self.layer_1(x))
        x = F.relu(self.layer_2(x))
        x = F.relu(self.layer_3(x))
        return self.layer_4(x)


# Recurrent neural network (many-to-one)
# https://github.com/yunjey/pytorch-tutorial/blob/master/tutorials/02-intermediate/recurrent_neural_network/main.py#L39-L58
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(RNN, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, num_classes)
    
    def forward(self, x):
        # Set initial hidden and cell states 
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device) 
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device)
        
        # Forward propagate LSTM
        out, _ = self.lstm(x, (h0, c0))  # out: tensor of shape (batch_size, seq_length, hidden_size)
        
        # Decode the hidden state of the last time step
        out = self.fc(out[:, -1, :])
        return out
```

```python
import torch, sys
from torch import nn
import numpy as np
#import pytorch_lightning as pl
from torch.nn import functional as F
from sklearn.model_selection import KFold
from torch.utils.data import DataLoader, random_split, ConcatDataset, Dataset
from tqdm import tqdm

use_cuda = torch.cuda.is_available()
device = torch.device("cuda" if use_cuda else "cpu")

input_tensor = torch.tensor(input_data)
input_tensor = torch.squeeze(input_tensor)
# print(len(input_tensor))
train_size = int((0.8* len(input_tensor)))
# print(train_size)
test_size = len(input_tensor) - train_size
# print(test_size)
# sys.exit()

# print(input_tensor.size())

# train_set = input_tensor[:train_size]
# print(len(train_set[0]))
# test_set = input_tensor[train_size+1:test_size]
# print(len(test_set))

train_set = torch.split(input_tensor[0], train_size)
# print(len(train_set[0]))
test_set = torch.split(input_tensor[0], int(len(input_data[0]))-train_size)
# print(len(test_set[0]))
# sys.exit()

# trainingData = torch.utils.data.DataLoader(train_set, batch_size=1, shuffle=True)
# testData = torch.utils.data.DataLoader(test_set, batch_size=1, shuffle=True)
# print(len(trainingData))
# print(input_tensor.size())
# print(input_tensor.shape)

# Configuration options
k_folds = 5
num_epochs = 50
loss_function = nn.L1Loss()

# For fold results
results = {}

# Set fixed random number seed
torch.manual_seed(42)


# dataset_train_part = trainingData
# dataset_test_part = testData
# dataset_train_part = input_tensor[0]
# print(dataset_train_part[:10])
# sys.exit()

# dataset_test_part = input_tensor[0]
# dataset = ConcatDataset([dataset_train_part, dataset_test_part])
# print(len(input_tensor))
# print(len(input_tensor[0]))
# print(input_tensor.shape)
# input_tensor = torch.squeeze(input_tensor)
# print(len(input_tensor))
# print(input_tensor.shape)
# sys.exit()
dataset = input_tensor
data_train, data_val = random_split(dataset, [train_size, test_size])

dataConcat = ConcatDataset([data_train, data_val])

print(len(data_train))
print(len(data_val))
# sys.exit()
# Define the K-fold Cross Validator
kfold = KFold(n_splits=k_folds, shuffle=True)
  
# Start print
print('--------------------------------')

# K-fold Cross Validation model evaluation
for fold, (train_ids, test_ids) in enumerate(kfold.split(dataConcat)):


  # print(len(train_ids))
  # print(len(test_ids))
  # sys.exit()
  
  # Print
  print(f'FOLD {fold}')
  print('--------------------------------')
  
  # Sample elements randomly from a given list of ids, no replacement.
  train_subsampler = torch.utils.data.SubsetRandomSampler(train_ids)
  test_subsampler = torch.utils.data.SubsetRandomSampler(test_ids)
  
  # Define data loaders for training and testing data in this fold
  trainloader = torch.utils.data.DataLoader(
                    dataConcat, 
                    batch_size=10, sampler=train_subsampler)
  print(len(trainloader))
  # sys.exit()
  testloader = torch.utils.data.DataLoader(
                    dataConcat,
                    batch_size=10, sampler=test_subsampler)
  
  # Init the neural network
  # network = SimpleConvNet()
  network = Fully_Connected(3,1)
  network.apply(reset_weights)
  network = network.to(device)

  # print('num params in model =', sum(parameter.numel() for parameter in network.parameters()))
  
  # Initialize optimizer
  optimizer = torch.optim.Adam(network.parameters(), lr=1e-4)
  
  # Run the training loop for defined number of epochs
  for epoch in tqdm(range(0, num_epochs)):

    # Print epoch
    print(f'Starting epoch {epoch+1}')

    # Set current loss value
    current_loss = 0.0

    # Iterate over the DataLoader for training data
    for i, data in tqdm(enumerate(trainloader, 0)):
      
      # Get inputs
      # print(data)
      # print(data.shape)
      # print(data[0])
      # sys.exit()
      # inputs = data[:][:3]
      input_indices = torch.tensor([0,1,2])
      target_indices = torch.tensor([3])
      # print(indices.shape)
      inputs = torch.index_select(data, 1, input_indices).to(device)
      targets = torch.index_select(data, 1, target_indices).to(device)
      # print(input)
      # print(target)
      # print(inputs.size())
      # targets = data[:][:][3]
      # print(inputs)
      # # print(targets)
      # print(inputs.shape)
      # print(targets.shape)
      # print(input.shape)
      # sys.exit()
      # inputs  = data
      
      # Zero the gradients
      optimizer.zero_grad()
      
      # Perform forward pass
      outputs = network(inputs)
      
      # Compute loss
      loss = loss_function(outputs, targets)
      # loss = loss_function(outputs)

      if i == len(trainloader)-1:
        print(loss)
      
      # Perform backward pass
      loss.backward()
      
      # Perform optimization
      optimizer.step()
      
      # Print statistics
      current_loss += loss.item()

    avg_loss = current_loss/len(trainloader)
    print('epoch avg loss = ', avg_loss)
          
  # Process is complete.
  print('Training process has finished. Saving trained model.')

  # Print about testing
  print('Starting testing')
  
  # Saving the model
  save_path = f'./model-fold-{fold}.pth'
  torch.save(network.state_dict(), save_path)

  # Evaluationfor this fold
  correct, total = 0, 0
  with torch.no_grad():

    # Iterate over the test data and generate predictions
    for i, data in enumerate(testloader, 0):

      # Get inputs
      input_indices = torch.tensor([0,1,2])
      target_indices = torch.tensor([3])
      inputs = torch.index_select(data, 1, input_indices).to(device)
      targets = torch.index_select(data, 1, target_indices).to(device)

      # Generate outputs
      outputs = network(inputs)

      # Set total and correct
      _, predicted = torch.max(outputs.data, 1)
      total += targets.size(0)
      correct += (predicted == targets).sum().item()

    # Print accuracy
    print('Accuracy for fold %d: %d %%' % (fold, 100.0 * correct / total))
    print('--------------------------------')
    results[fold] = 100.0 * (correct / total)
  
# Print fold results
print(f'K-FOLD CROSS VALIDATION RESULTS FOR {k_folds} FOLDS')
print('--------------------------------')
sum = 0.0
for key, value in results.items():
  print(f'Fold {key}: {value} %')
  sum += value
print(f'Average: {sum/len(results.items())} %')
```
